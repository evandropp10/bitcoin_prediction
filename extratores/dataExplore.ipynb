```python
def fact_class_change(x):
    if x == "down4":
        return 1
    if x == "down3":
        return 2
    if x == "down2":
        return 3
    if x == "down1":
        return 4
    if x == "up1":
        return 5
    if x == "up2":
        return 6
    if x == "up3":
        return 7
    if x == "up4":
        return 8
    
df['fact'] = df['class_change_tomorrow'].apply(fact_class_change)
    
```

```python
plt.subplots(figsize=(15,6))
plt.hist(df['fact'])
```
